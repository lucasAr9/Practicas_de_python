```python
import urllib.request

# en url guardamos la url del sitio que queremos que abra el programa
# url = 'https://github.com/numpy/numpy/blob/main/README.md'
url = 'https://raw.githubusercontent.com/numpy/numpy/main/README.md'

# con request.urlopen(url) abrimos la url especificada
url_open = urllib.request.urlopen(url)

# en respuesta guardo el html de la url
leer_str = str(url_open.read().decode('utf-8'))
lineas = list(leer_str.split('href="' and '"' and '\n' and '<!--' and '(' and ')'))

# filtro informacion del html guardado en la variable los_http
los_http = list(filter(lambda lineas: 'http' in lineas, lineas))

for i in los_http:
    print(i)
```

```python
# 7.
frase = [
    'cruzamiento',
    'centrifugados',
    'portón',
    'casa',
    'día de sol',
    'con diez uñas',
    'no-se-duplica'
]
import string
diccionario = string.ascii_letters

# a)
for palabra in frase:
    pal = palabra.lower().replace("-", "").replace(" ", "")
    print(len(set(pal)) == len(pal))

# b)
for palabra in frase:
    palabra.lower()
    palabra = [letra for letra in palabra if letra in diccionario]
    print(len(set(palabra)) == len(palabra))
```
